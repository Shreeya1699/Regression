```python
import numpy as np  #importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
# import math
plt.rcParams['figure.figsize'] = (30.0, 20.0)
from mpl_toolkits.mplot3d import Axes3D

data=pd.read_csv(r"C:\Users\shreya\Downloads\data.csv", header=None)
# data=(data-data.mean())/data.std()
data= ((data-data.min())/data.max()-data.min())

data=np.asarray(data) #dataset to numpy array
data=data[:,1:]
data=np.asarray(data) #dataset to numpy array
y=data[:,2]
y = np.reshape(y, (len(y),1))

input=np.concatenate((np.ones((len(y),1),dtype=int),(data[:,:2])),axis=1) 

#concatenating a column of ones to the data
w= np.matmul(np.matmul(np.linalg.inv(np.matmul(np.transpose(input),input)),np.transpose(input)),y)

#Vectorized weight update rule

w=np.reshape(w,(1,3))
print(w)
e=np.matmul(w,np.transpose(input))
N=y.size
error=(np.subtract(np.transpose(y),e))
RMS=((np.sum(np.square(error)))/N)**0.5
SS_tot= np.sum(np.square(y))
SS_res= np.sum(np.square(error))
R2= 1- (SS_res/SS_tot)
print("RMSE:",RMS)
print("R2 value:", R2)
```
